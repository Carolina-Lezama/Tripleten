```python
#importar dataframe para hipotesis
df_7_sql = pd.read_csv('/datasets/project_sql_result_07.csv')
# start_ts: fecha y hora de la recogida
# weather_conditions: condiciones clim√°ticas en el momento en el que comenz√≥ el viaje
# duration_seconds: duraci√≥n del viaje en segundos
```

```python
#revisar los datos que contienen
df_7_sql.info()
print()
print(df_7_sql.sample(5))
```

observaciones:
este cuenta con 1068 filas, contiene 3 columnas, ninguna de estas columnas posee datos nulos, las segunda columna tiene bien el tipo de dato, podriamos pasar la primera columna a formato datatime,, y en la ultima verificar si es necesario el punto decinal o es posible cambiarlo a int

```python
#estudiar los datos que contienen
# ver deciamles para duration_seconds
print("Ver valores")
print(df_7_sql['duration_seconds'].sample(10))

# verificar decimales
print((df_7_sql['duration_seconds'] % 1 != 0).sum())

# Redondear hacia arriba y convertir a entero
df_7_sql['duration_seconds'] = np.ceil(df_7_sql['duration_seconds']).astype(int)

print("Ver valores nuevamente")
print(df_7_sql['duration_seconds'].sample(10))
```

```python
#cambiar de object a datatime
df_7_sql['start_ts'] = pd.to_datetime(df_7_sql['start_ts'])

df_7_sql['dia_semana'] = df_7_sql['start_ts'].dt.weekday
df_7_sql['sabado'] = df_7_sql['start_ts'].dt.weekday == 5

df_7_sql['hora'] = df_7_sql['start_ts'].dt.hour
df_7_sql['fecha'] = df_7_sql['start_ts'].dt.date

df_7_sql.info()
print(df_7_sql.sample(10))
```

peque√±o recordatorio: /datasets/project_sql_result_07.csv contiene datos sobre viajes desde el Loop hasta el Aeropuerto Internacional O'Hare. 

```python
#declaracion de alpha
alpha = .05

#verificar dia de la semana, todos son sabados
print(df_7_sql['dia_semana'].unique())

# Filtrar s√°bados lluviosos
sabados_lluviosos = df_7_sql[
    (df_7_sql['weather_conditions'] == 'Bad')
]

# filtrar dias normales
sabados_normales = df_7_sql[
      (df_7_sql['weather_conditions'] == 'Good')
]
print(sabados_lluviosos.head())
print('-'*30)
print(sabados_normales.head())
```

h0:"La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare no cambia los s√°bados lluviosos"
h1:"La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los s√°bados lluviosos"

```python
results = st.ttest_ind(sabados_lluviosos['duration_seconds'], sabados_normales['duration_seconds']) # si el valor p es menor que alpha, rechazamos la hip√≥tesis
print('valor p: ', results.pvalue) 
if results.pvalue < alpha: # comparar el valor p con el umbral
    print("Rechazamos la hip√≥tesis nula")
    print("La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los s√°bados lluviosos")
else:
    print("No podemos rechazar la hip√≥tesis nula")
    print("La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare no cambia los s√°bados lluviosos")
```

```python
# Compararemos ambos dias
print("sabados lluviosos")
print(f"Cantidad de viajes: {len(sabados_lluviosos)}")
print(f"Duraci√≥n promedio: {sabados_lluviosos['duration_seconds'].mean():.2f}")
print(f"Desviaci√≥n est√°ndar: {sabados_lluviosos['duration_seconds'].std():.2f}")
print(f"Mediana: {sabados_lluviosos['duration_seconds'].median()}")
print()
print("sabados normales")
print(f"Cantidad de viajes: {len(sabados_normales)}")
print(f"Duraci√≥n promedio: {sabados_normales['duration_seconds'].mean():.2f}")
print(f"Desviaci√≥n est√°ndar: {sabados_normales['duration_seconds'].std():.2f}")
print(f"Mediana: {sabados_normales['duration_seconds'].median()}")
print()
print(df_7_sql['weather_conditions'].value_counts())
print(f"Porcentaje de d√≠as lluviosos: {len(sabados_lluviosos)/len(df_7_sql)*100:.2f}%")
```

Revision de lo anterior:
podemos ver que efectivamente hay un cambio en los promedios, de si es normal o nublado, siendo una diferencia bastante notable, podemos ver que los numeros crecen enormemente ya que la diferencia en la cantidad de viajes es abismal, y aun logran subir bastante el promedio, tambien hay diferencia en la desviacion estandar, y podemos confirmar todo esto tambien con la mediana ya que hay casi 700 segundos de diferencia.
no olvidemos que solo el 16.8% de todas las filas pertenecen a dias lluviosos

C√≥mo plante√© las hip√≥tesis:
use de referencia el mensaje indicadado "La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los s√°bados lluviosos"

La hip√≥tesis nula (H‚ÇÄ) siempre implica ausencia de diferencias.
La hip√≥tesis alternativa (H‚ÇÅ) contradice la nula, afirmando que S√ç hay diferencia entre los grupos.

Por tanto:
- H‚ÇÄ: No hay diferencias en los promedios de duraci√≥n
- H‚ÇÅ: S√≠ hay diferencias significativas en los promedios

use la hip√≥tesis sobre la igualdad de las medias de dos poblaciones:
ya que son 2 muestras diferentes y independientes, tengo una variable numerica continua que son los segundosm especificamente se me pedia comprar medias(promedios) y saber si estos cambian entre ambos grupos, mis muestras son lo suficientemente grandes para usar esto

elegi 0.05 para el nivel de significancia ta qye este es el estandar usado en alasisis estadisticos, acepto un 5% de probailidad de error
como mi valor p es extremadamente peque√±o, entonces rechazo la hipotesis nula ya que hay suficiente evidencia estadistica para decir que los promedios si cambian de un dia lluvioso a uno normal
